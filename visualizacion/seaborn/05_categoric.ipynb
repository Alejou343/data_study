```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.countplot(data=tips, y='day', hue='sex')
plt.show()

# This diagram shows total tips provided by male and female people along the days
# The friday is the day with less tips
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips, x='day', hue='sex', y='total_bill')
plt.show()

# Another kind of graph used to ubicate points clasified by days and sex
```

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, x='day', hue='sex', y='total_bill')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', hue='sex', y='total_bill')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', hue='sex', y='total_bill')
sns.swarmplot(data=tips, x='day', hue='sex', y='total_bill', dodge=True, palette='light:0.2534534', marker='<')
plt.show()

# dodge to separate by a categorical variable
# palette to apply a theme of colors to points o parts of graph
# marker to apply a new icon different to dots 
```

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x='day', hue='sex', y='total_bill', palette='pastel', split=True)
plt.show()

# Show the normal distribution by a categorical variable (day) and (sex)
# Split is used to join values by a categorical variable (sex in this case)
```

```python
sns.catplot(data=tips, x='day', hue='sex', y='total_bill', dodge=True, col='time', kind='strip')
plt.show()

# Catplot is used as a generic graph if you set the propertie kind as string, the graph change their class
# col add another categorical variable 
# dodge group data by a categorical variable
```
