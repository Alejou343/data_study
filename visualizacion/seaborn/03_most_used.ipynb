```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()                      # --> Another way to show a dataframe
```

```python
sns.plot
```

```python
sns.displot(data=tips, x='total_bill')      # Frequency histogram using variable total_bill and y means frequency of every single x
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip')      # Frequency histogram using variable total_bill and y means frequency of every single x
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='day', kind='kde', legend=True, palette='dark', alpha=0.75)
plt.show()

# This graph contains several properties
# x to indicate horizontal axis values
# y to indicate vertical axis values (in many cases the frequency)
# hue to separate data by a propertie (sex in this case)
# kind to change kind of graph (to density in this case)
# Legend to show right legend description
# palette to change colors of graph
# alpha to change colors transparency
```
