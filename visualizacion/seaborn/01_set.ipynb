```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
sns.barplot(x=['A', 'B', 'C'], y=[1,3,2])           # You need an x axis information and y axis values used arrays
plt.show()                                          # Can we use matplotlib syntax
```

```python
sns.set(style='darkgrid', palette='dark', font="Sans", font_scale=2)    # Set is used to provide styles to a plot (style= lienz, palette=color_intensity, font= kind letter)
sns.barplot(x=['A', 'B', 'C'], y=[1,3,2])
plt.show()
```
