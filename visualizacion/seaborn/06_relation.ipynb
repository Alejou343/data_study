```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
plt.figure(figsize=(10,10))
markers = {'Lunch': 'D', 'Dinner': 's'}
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()

# This graph show the relationship between tip value and total bill value
```

```python
# Another kind of graph
sns.lineplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', kind='scatter', col='smoker')
# fig, axes = plt.subplots(2,2, figsize=(10,10))
# To use subplots --> sns.scatterplot(ax = axes[1,0], data, ...)
plt.show()

# kind is used to apply a new kind of graph
# col is used to separate in columns by a variable
```

```python
sns.pairplot(data=tips, hue='sex')
plt.show()

# This kind of plot is to see differents graphs between two variables
```
