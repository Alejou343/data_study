```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
# Show two subplots --> 1 at left and another at right
plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,4)
plt.hist(y)
plt.show()
```

```python
# Show 2 subplots --> one at left and another at right with some modifications in formats
plt.subplot(1,2,1)
plt.plot(x,y, 'ro-')
plt.plot(y,x, 'go-')

plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

```python
# Show 2 subplots one at top and another at bottom
plt.subplot(2,1,1)
plt.plot(x,y, 'ro-')
plt.plot(y,x, 'go-')

plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```
