```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# View available plots styles
print(plt.style.available)
```

```python
# Data to plot
country = ['INDIA', 'JAPON', 'MEXICO', 'COLOMBIA', 'ALEMANIA']
population = [1000, 800, 900, 1000, 300]
colors = ['#8ECAE6', '#219EBC', '#023047', '#FFB703', '#FB8500']
```

```python
# Graph of bar plot

plt.bar(country, population, width=0.5, color=colors)                 # --> width controls the line width, colors receive an array with color information
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'), rotation=45)  # --> This method is used to modify x values, rotation to rotare labels
plt.title('Population by country')
plt.xlabel('Country')
plt.ylabel('x 10^3 people')
plt.show()
```

```python
# Horizontal bar graph
plt.barh(country, population, height=0.5, color=colors)                 # --> width controls the line width, colors receive an array with color information
plt.title('Population by country')
plt.xlabel('x 10^3 people')
plt.ylabel('Country')
plt.show()
```
