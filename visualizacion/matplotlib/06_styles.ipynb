```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10,10,200)
```

```python
# Show all styles provides by matplotlib.pyplot
print(plt.style.available)
```

```python
# Use any style provided by the library
plt.style.use('dark_background')
```

```python
# This method create plots in the same graph --> Styles with Matlab method
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x, x+1, 'r--')
ax.plot(x, x+2, 'b-')
ax.plot(x, x+3, 'y.-')
ax.plot(x, x+4, 'go:')
```

```python
plt.style.use('seaborn-v0_8-dark-palette')
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x, x+1, color="blue", alpha=0.4, linewidth=2, linestyle='-', marker='p', markersize=10)             # Alpha to set transparency of the line
ax.plot(x, x+2, color="#66FF66", alpha=1.0, linewidth=3, linestyle='--', marker='x')                        # Color can be name of color or hex code
ax.plot(x, x+3, color="cyan", alpha=1.0, linewidth=1, linestyle='dashed', marker='v')                       # Linewidth 1o control line grossor
ax.plot(x, x+4, color="#CCFFCC", alpha=1.0, linewidth=1, linestyle='-', marker='p')                         # marker to define shape in each point of graph
```

```python
for i in plt.style.available:
    plt.style.use(i)
    fig, axes = plt.subplots(figsize=(8,8))
    axes.plot(x, np.cos(x))
    axes.set_title(i)
```
