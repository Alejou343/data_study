```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create an array with data
data = np.random.randint(1,50,100)
data
```

```python
plt.hist(data, bins=50, color="green")
plt.show()
```

```python
# Another kind of graph
plt.hist(data, bins=10, histtype='step', color="red")
plt.show()
```

```python
# Boxplot --> this kind of graph show percentile 25, 50, 75 and 100
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True)    # vert: Rotate graph to horizontal, patch_artist: Apply color and notch reduce median area
plt.show()
```

```python
# Scatter --> First step generate values to relation between two of them
N = 50
x = np.random.rand(N)                   # X values correspond to a horizontal position
y = np.random.rand(N)                   # Y values correspond to a vertical position
area = (30 * np.random.rand(N)) ** 2    # Area as third parameter correspond to a different ball size
colors = np.random.rand(N)              # Every single value correspond to a different color
```

```python
plt.scatter(x,y, s=area, c=colors, marker='o', alpha=0.5)
plt.show()
```
