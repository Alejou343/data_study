```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5,5,11)
y = x ** 2
```

```python
# Oriented object method
fig = plt.figure()                          # Lienzo
axes = fig.add_axes([0.1,0.1,0.5,0.9])      # add_axes([positionx, positiony, sizex, sizey])
axes2 = fig.add_axes([0.7,0.1,0.5,0.9])
axes.plot(x,y, 'b')                         # Graphic in axes1
axes2.plot(x,y, 'ro-')                      # Graphic in axes2
axes2.set_facecolor('gray')                 # set_facecolor to chage the background color
plt.show()
```

```python
# The same method oriented object but using a function 
def graphic(x,y,color):
    fig = plt.figure()
    axes = fig.add_axes([0.1,0.1,0.5,0.9])
    axes2 = fig.add_axes([0.7,0.1,0.5,0.9])
    axes.plot(x, y, 'b')
    axes2.plot(x, y, 'go')
    axes2.set_facecolor(color)
    return plt.show()

graphic(x,y,'gray')
```
