```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a list of numbers
x = np.linspace(-5,5,40)

# Create a function to relation both variables
y = x**2 - 2 * x + 4
```

```python
# Graphic of a function to reveal a correlation between two variables
plt.plot(x,y)
plt.show()
```

```python
# The same kind of function but their format has been changed by dots
plt.plot(x,y, 'ro')     # --> Color red and o to show values in dots
plt.show()
```

```python
# The same kind of function but their format has been changed by 'x' with a continous line
plt.plot(x,y, 'yx-')     # --> plot with yellow cross
plt.show()
```

```python
# The same kind of function but their format has been changed by squares with a continous line
plt.plot(x,y, 'bs-')
plt.show()
```

```python
# The same kind of function but their format has been changed by squares with a line of dots
plt.plot(x,y, 'gD:')
plt.show()
```

```python
# This kind of graphic shows the correlation between two variables when the height of bars is provided by a function
plt.bar(x,y)
plt.show()
```

```python
# This kind of graphic is an hystogram of frequencys
plt.hist(y)
plt.show()
```

```python
# This kind of graphic show a percentage of any variable in a total that is the total sum of all values
plt.pie(y)
plt.show()
```

```python
# This kind of graphic shows a distribution of dots that mean the correlation between two variables
plt.scatter(x,y)
plt.show()
```

```python
# This kind of graphic show the average and standard deviation of these values
plt.boxplot(x)
plt.show()
```

```python
# This diagram is very similar to bar diagram
plt.stem(x,y)
plt.show()
```

```python
# This kind of graphic shows the area under the line provided by function
plt.stackplot(x,y)
plt.show()
```

```python
plt.stairs(x)
plt.show()
```
