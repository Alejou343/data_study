```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define x and y values
x = np.linspace(-10, 10, 1000)
y = np.cos(x)
```

```python
# Plot the results 
fig, axes = plt.subplots(1,2, figsize=(5,5))    # Figsize to control the size of plot

# Left plot
axes[0].plot(x,y,label="$sin(x)$")              # Label to show tags into a plot
axes[0].set_title("f(x) = sin(x)")              # Title of a plot
axes[0].set_xlabel("X values")                  # Name of x label
axes[0].set_ylabel("Y values")                  # Name of y label
axes[0].legend(loc="upper right")               # Tag location 


# Right plot
axes[1].plot(y,x, label="$sin(y)$")
axes[1].set_title("f(y) = sin(y)")
axes[1].set_xlabel("Y values")
axes[1].set_ylabel("X values")
axes[1].legend()

# Correction of spacing
fig.tight_layout()
```

```python
# Non oriented object method

plt.plot(x,y, label="$cos(x)$")                 # Label to show a tag into plot
plt.title('$f(x) = cos(x)$')                    # Title of plot
plt.xlabel('$x$')                               # X label name
plt.ylabel('$cos(x)$')                          # Y label name
plt.legend(loc = "upper right")                 # Position of tag
```
