```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create bot arrays to graph
x = np.linspace(-10,10, 1000)
y = np.sin(x)
```

```python
# A way to make two subplots
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y, 'b')
axes[1].plot(y,x, 'g')
plt.show()
```

```python
# Another way with the same result
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b')
ax2.plot(y,x, 'g')
plt.show()
```

```python
# Example using 8 plots (4 in the first row and 4 in the second row)
fig, axes = plt.subplots(2,4)
axes[0, 0].plot(x, np.sin(x), 'r-.')
axes[0, 1].plot(x, np.exp2(x) ** - 1, 'y:')
axes[0, 2].plot(x, np.tan(x), 'g')
axes[0, 3].plot(x, x**2 - 2*x + 1, 'b')
axes[1, 0].plot(x, np.sqrt(x**2), 'y:')
axes[1, 1].plot(x, 1/x, 'r')
axes[1, 2].plot(x, 2 * x + 1, 'y')
axes[1, 3].plot(x, 4 ** x - 10, 'g-.')
fig.tight_layout()                  # This function clean the plot using a responsive layout
```

```python
# Another way using tuples
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2,4)
ax1.plot(x, np.sin(x))
ax2.plot(x, np.exp2(x) ** - 1)
ax3.plot(x, np.tan(x))
ax4.plot(x, x**2 - 2*x + 1)
ax5.plot(x, np.sqrt(x**2))
ax6.plot(x, 1/x)
ax7.plot(x, 2 * x + 1)
ax8.plot(x, 4 ** x - 10)
fig.tight_layout()                  # This function clean the plot using a responsive layout
```

```python
# Using functions 

def grapheight(x):
    fig, axes = plt.subplots(2,4)
    axes[0,0].plot(x, np.cos(x))
    axes[0,1].plot(x, np.sin(x))
    axes[0,2].plot(x, np.tan(x))
    axes[0,3].plot(x, np.tanh(x))
    axes[1,0].plot(x, np.absolute(x))
    axes[1,1].plot(x, np.arcsin(x))
    axes[1,2].plot(x, np.cosh(x))
    axes[1,3].plot(x, x**2 - 2*x + 1)
    fig.tight_layout()
    
x = np.linspace(-10,10, 1000)

grapheight(x)
```
