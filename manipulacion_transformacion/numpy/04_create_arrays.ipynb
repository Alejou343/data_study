```python
import numpy as np
```

```python
# Create array using python syntax

lista_py = list(range(0,10))
array_from_py = np.array(lista_py)
array_from_py
```

```python
# Create array using numpy sintax

array_from_np = np.arange(0,21, 5, dtype='int64')
array_from_np
```

```python
# Create array with full zeros

zero_array = np.zeros(10)
print(zero_array)
zero_array = np.zeros([10, 5])
print(zero_array)
```

```python
# Create array with full ones

one_array = np.ones(10)
print(one_array)
one_array = np.ones([10, 5])
print(one_array)
```

```python
# Create an array with 400 values between 0 and 10

range_arr = np.linspace(0, 10, 400)
range_arr
```

```python
# Create an identity matriz with this dimensions 4 columns and 4 rows

identidad = np.eye(4,4)
identidad
```
